# 一元线性回归

```python
#加载数据
x=np.array([137.97,104.50,100.00,124.32,79.20,99.00,124.00,114.00,106.69,138.05,53.75,46.91,68.00,63.02,81.26,86.21])
y=np.array([145.00,110.00,93.00,116.00,65.32,104.00,118.00,91.00,62.00,133.00,51.00,45.00,78.50,69.65,75.69,95.30])
```

```python
#设置超参数
learn_rate=0.0001
iter=10
display_step=1
```

```python
#设置模型参数初值
np.random.seed(612)
w=np.random.randn()#randn函数返回一个或一组样本，具有标准正态分布
b=np.random.randn()
```

```python
#训练模型
mse=[]

for i in range(0,iter+1):
    
    dL_dw=np.mean(x*(x*w+b-y))
    dL_db=np.mean(x*w+b-y)
    w=w-learn_rate*dL_dw
    b=b-learn_rate*dL_db
    
    pred=w*x+b
    Loss=np.mean(np.square(y-pred))/2
    mse.append(Loss)
    
    if i%display_step==0:
        print("i:%i, Loss:%f, w:%f, b:%f "%(i,mse[i],w,b))
```

```python
#结果可视化
plt.figure()

plt.scatter(x,y,color="r",label="销售记录")
plt.scatter(x,pred,color="b")
plt.plot(x,pred,color="b",label="梯度下降法")

plt.xlabel("Area",fontsize=15)
plt.ylabel("Price",fontsize=15)

plt.legend(loc="upper left")
plt.show()
```

```python
# 解析法对比
plt.figure()

plt.scatter(x,y,color="r",label="销售记录")
plt.scatter(x,pred,color="b")
plt.plot(x,pred,color="b",label="梯度下降法")
plt.plot(x,0.8945604*x+5.4108505,color="g",label="解析法")#

plt.xlabel("Area",fontsize=15)
plt.ylabel("Price",fontsize=15)

plt.legend(loc="upper left")
plt.show()
```

# 多元线性回归

```python
#加载数据
area=np.array([137.97,104.50,100.00,124.32,79.20,99.00,124.00,114.00,106.69,138.05,53.75,46.91,68.00,63.02,81.26,86.21])
room=np.array([3,2,2,3,1,2,3,2,2,3,1,1,1,1,2,2])
price=np.array([145.00,110.00,93.00,116.00,65.32,104.00,118.00,91.00,62.00,133.00,51.00,45.00,78.50,69.65,75.69,95.30])
num=len(area)

x0=np.ones(num)
#线性归一化
x1=(area-area.min())/(area.max()-area.min())
x2=(room-room.min())/(room.max()-room.min())
x0,x1,x2
```

```python
X=np.stack((x0,x1,x2),axis=1)
Y=price.reshape(-1,1)
print(Y)
X.shape,Y.shape
```

```python
#设置超参数
learn_rate=0.01
iter=500
display_step=50
#设置模型参数初始值
np.random.seed(612)
W=np.random.randn(3,1)#randn函数返回一个或一组样本，具有标准正态分布
```

```python
#训练模型
mse=[]

for i in range(0,iter+1):
    
    PRED=X@W
    #PRED=np.matmul(X,W)
    Loss=np.mean(np.square(Y-PRED))/2
    mse.append(Loss)
    
    dL_dW=X.T@(X@W-Y)
    #dL_dW=np.matmul(np.transpose(X),np.matmul(X,W)-Y)
    W=W-learn_rate*dL_dW

    if i%display_step==0:
        print("i:%i, Loss:%f "%(i,mse[i]))
```

```python
plt.figure(figsize=(12,4))

plt.subplot(121)
plt.plot(mse)
plt.xlabel("Iteration",fontsize=15)
plt.ylabel("Loss",fontsize=15)

plt.subplot(122)
PRED=PRED.reshape(-1)
print(PRED)
plt.plot(price,color="r",label="销售记录")
plt.plot(PRED,color="b",label="梯度下降法numppy")

plt.legend()
plt.show()
```

```python
#自动求导
#with tf.GradientTape() as tape:
#    y=
#dy_dx=tape.gradient(y,x)

x=tf.Variable(3.)
with tf.GradientTape() as tape:
    y=tf.square(x)
dy_dx=tape.gradient(y,x)
print(y)
print(dy_dx)
```

```python
#GradientTape(persistent,watch_accessed_variables) 
x=tf.Variable(3.)
with tf.GradientTape(persistent=True) as tape:
    y=tf.square(x)
    z=pow(x,3)
dy_dx=tape.gradient(y,x)
dz_dx=tape.gradient(z,x)
print(y)
print(dy_dx)
print(z)
print(dz_dx)

del tape
```

```python
x=tf.Variable(3.)
with tf.GradientTape(watch_accessed_variables=False) as tape:
    y=tf.square(x)
dy_dx=tape.gradient(y,x)
print(y)
print(dy_dx)
```

```python
#添加监视
#watch()
x=tf.Variable(3.)
with tf.GradientTape(watch_accessed_variables=False) as tape:
    tape.watch(x)
    y=tf.square(x)
dy_dx=tape.gradient(y,x)
print(y)
print(dy_dx)
```

```python
#监视非可训练变量
x=tf.constant(3.)
with tf.GradientTape(watch_accessed_variables=False) as tape:
    tape.watch(x)
    y=tf.square(x)
dy_dx=tape.gradient(y,x)
print(y)
print(dy_dx)
```

```python
#多元函数求偏导
#tape.gradient(函数，自变量)
x=tf.Variable(3.)
y=tf.Variable(4.)

with tf.GradientTape() as tape:
    f=tf.square(x)+2*tf.square(y)+1
    
df_dx,df_dy=tape.gradient(f,[x,y])

print(f)
print(df_dx)
print(df_dy)
```
