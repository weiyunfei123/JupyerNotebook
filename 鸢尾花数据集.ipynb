```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

URL="http://download.tensorflow.org/data/iris_training.csv"
URL_list=URL.split("/")
fname=URL_list[-1]
origin=URL
train_path=tf.keras.utils.get_file(fname,origin)

plt.rcParams["font.sans-serif"]=["SimHei"]
plt.rcParams["axes.unicode_minus"]=False

column=["花萼长度","花萼宽度","花瓣长度","花瓣宽度","种类"]
df_iris=pd.read_csv(train_path,header=0,names=column)
iris=np.array(df_iris)

plt.figure(figsize=(5,5)) 
plt.scatter(iris[:,2],iris[:,3],c=iris[:,4],cmap="brg")
plt.title("鸢尾花数据集（蓝色->山鸢尾花|红色->变色鸢尾花|绿色->维基利亚鸢尾花）")
plt.xlabel(column[2])
plt.ylabel(column[3])
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

URL="http://download.tensorflow.org/data/iris_training.csv"
URL_list=URL.split("/")
fname=URL_list[-1]
origin=URL
train_path=tf.keras.utils.get_file(fname,origin)

plt.rcParams["font.sans-serif"]=["SimHei"]
plt.rcParams["axes.unicode_minus"]=False

column=["花萼长度","花萼宽度","花瓣长度","花瓣宽度","种类"]
df_iris=pd.read_csv(train_path,header=0,names=column)
iris=np.array(df_iris)

plt.figure(figsize=(15,15))
for i in range(4):
    for j in range(4):
        plt.subplot(4,4,4*i+j+1)
        #if(i==j):
            #plt.text(0.35,0.5,column[i],fontsize=15)#设置文本标题
        #else:
        plt.scatter(iris[:,i],iris[:,j],c=iris[:,4],cmap="brg")
        plt.xlabel(column[i])
        plt.ylabel(column[j])   
        plt.title(str(4*i+j+1)+"."+column[i]+"-"+column[j])
plt.tight_layout()
plt.suptitle("鸢尾花数据集（蓝色->山鸢尾花|红色->变色鸢尾花|绿色->维基利亚鸢尾花）",x=0.5,y=1.05,fontsize=20)
plt.show()
            
```
